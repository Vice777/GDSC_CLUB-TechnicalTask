## What is Multi-Layer Perceptron (MLP)

<img src="https://miro.medium.com/max/828/1*-IPQlOd46dlsutIbUq1Zcw.png">


<p style="text-align: justify;">A multilayer perceptron (MLP) is a fully connected class of feedforward artificial neural network (ANN). The term MLP is used ambiguously, sometimes loosely to mean any feedforward ANN, sometimes strictly to refer to networks composed of multiple layers of perceptrons (with threshold activation).</p>

<p style="text-align: justify;">An MLP consists of at least three layers of nodes: an input layer, a hidden layer and an output layer. Except for the input nodes, each node is a neuron that uses a nonlinear activation function. MLP utilizes a supervised learning technique called backpropagation for training.Its multiple layers and non-linear activation distinguish MLP from a linear perceptron. It can distinguish data that is not linearly separable.</p>

## Working of MLP

The Basic steps are as follows:<br>
<ol>
    <li>Initialize the input Weights and Bias for hidden and output layer </li>
    <li>Forward Propogation and selection of activation layer </li>
    <li>Backward Propogation for updating weights and bias </li>
</ol>

### Step 1: Initializing Weights

<b>For N-features in X_lable and N_hidden_layer:<b><br>
$ \text{W} = \text{range: }[(\frac{-1}{\sqrt{\text{N-features}}}), (\frac{1}{\sqrt{\text{N-features}}})] \text{  for N_hidden_Layer}  $ <br>
$ \text{W}_o = \text{[0,0,0, ..... n] , where n = N_hidden_layer}$ <br>

$ \text{V} = \text{range: }[(\frac{-1}{\sqrt{\text{N_hidden_layer}}}), (\frac{1}{\sqrt{\text{N_hidden_layer}}})] \text{  for Output Features}  $ <br>
$ \text{V}_o = \text{[0,0,0, ..... n] , where n = Output Features}$ <br> 

Where,<br>
$ \text{W} $   - Hidden Weight <br>
$ \text{W}_o$ - Bias In hidden layer <br>
$ \text{V} $   - Output Weight <br>
$ \text{V}_o$ - Bias in Output Weight <br>

### Step 2: Forward propagation

In order to proceed we need to improve the notation we have been using. That for, for each layer $1\geq l\geq L$, the activations and outputs are calculated as:

$$
\text{a}^l_j = {\sum^n_{i=0} w^l_{ji} x^l_i\,+ b^i_h = w^l_{j,0} x^l_0 + w^l_{j,1} x^l_1 + w^l_{j,2} x^l_2 + ... + w^l_{j,n}} x^l_n + b^i_h 
$$
$$Y^l_j = g^l(\text{a}^l_j)\,,$$
$$
    \text{a}^l_{j+1} = {\sum^n_{i=0} w^l_{j+1i} x^l_i\,+ b^i_o = w^l_{j+1,0} a^l_0 + w^l_{j+1,1} a^l_1 + w^l_{j+1,2} a^l_2 + ... + w^l_{j+1,n}} a^l_n + b^i_o
$$
$$\hat{Y}^l_j = g^l(\text{a}^l_{j+1})\,,$$
where:

<li> $y^l_j$ : $j^{th}$ output of layer $l$
<li> $\hat{y}^l_j$ : Final Output Layer 
<li> $x^l_i$ : $i^{th}$ input to layer $l$ 
<li> $w^l_{ji}$ : Weight of the $j^{th}$ neuron connected to $i^{th}$ input
<li> $b^i_h$ : Bias of the $i^{th}$ hidden layer
<li> $b^i_o$ : Bias of the $i^{th}$ output layer
<li> $\text{a}^l_j$ : Net activation
<li> $g^l()$  : Activation function of layer $l$.

### Activation Functions


<ol>
    <li>
        <b>Sigmoid Function </b> $$ \sigma(z)= \frac{1}{1+\exp^{-z}}$$
    </li> <br>
    <li>
        <b>Gradient Sigmoid Function </b>$$ \frac{\partial(\sigma(z))}{\partial(z)}= \sigma(z)*(1-\sigma(z)) $$ 
    </li><br>
    <li>
        <b>SoftMax Function </b>$$ \alpha(\vec{z})_i= \frac{\text{e}^{z_i}}{\sum^k_{j=1}\text{e}^{z_i}}$$ 
    </li><br>
    <li>
        <b>Gradient SoftMax Function </b>$$ \frac{\partial(\alpha(z))}{\partial(z)}= \alpha(z)*(1-\alpha(z)) $$ 
    </li><br>
    <li>
        <b>Cross-Entropy Loss </b>$$ \delta(\hat{y},y)= -\text{y}*\log{(\hat{y})}-(1-\text{y}) * \log{(1-{\hat{y}})}$$ 
    </li><br>
    <li>
        <b>Gradient Cross-Entropy Loss</b>$$ \delta(\hat{y},y)= -(\frac{\hat{y}}{y})+ \frac{(1-\hat{y})}{(1-y)} $$  
    </li><br>
</ol>

# Backward Propogation
<br>

Step 1 : Gradient loss of output Layer $(L_n)$ with respect to input layer $(l_{n-1})$:            $\text{grad}^{out}_{input} = \delta(\hat{y},y) * \sigma(\text{a}_{j-1})) $
<br>
Step 2:  Gradient of output = hidden_output . $(\text{grad}^{out}_{in})$ :
 $\text{grad}_{out} = \text{a}_j^T . \text{grad}^{out}_{in}  $
<br>

Step 3: Gradient of bias of output layer : $\text{grad}_{bo} = \sum(\text{grad}^{out}_{in}) $
<br>


Step 4 : Gradient loss of input Layer $(L_n)$ with respect to ouput layer $(l_{n-1})$:
 $\text{grad}^{in}_{out} = \text{grad}^{out}_{in} . {V^T} * \frac{\partial(\alpha(a_j))}{\partial(a_j)} $
<br>

Step 5:  Gradient of input layer :
 $\text{grad}_{in} = \text{X}^T . \text{grad}^{in}_{out}  $
<br>

Step 6: Gradient of bias of input layer : 
 $\text{grad}_{bi} = \sum(\text{grad}^{in}_{out}) $
<br>

Step 7: Updating Weights : 

<li>${V} = {V} - \text{learning rate} * \text{grad}_{out} $
<li>${V_o} = {V_o} - \text{learning rate} * \text{grad}_{bo} $
<li>${W} = {W} - \text{learning rate} * \text{grad}_{in} $
<li>${W_o} = {W_o} - \text{learning rate} * \text{grad}_{bi} $


<br>

## Implementing Multi Layer Perceptron from Scratch

```python
class MultiLayerPerceptron:
     
    def __init__(self,hidden_layer, epoch, learning_rate, verbose=False):
        self.hidden_layer = hidden_layer
        self.epoch = epoch
        self.learning_rate = learning_rate
        self.verbose = verbose
        
    # Initializing the weights    
    def initial_weights(self, X, y):
        n_sample, n_features = X.shape
        n_output = y.shape[1]
        
        limit_hidden = 1.0/math.sqrt(n_features)
        self.hiddenWeight = np.random.uniform(-limit_hidden,limit_hidden, (n_features, self.hidden_layer))        
        self.BiasHidden = np.zeros((1,self.hidden_layer))
        
        limit_out = 1.0/ math.sqrt(self.hidden_layer)
        self.outputWeight = np.random.uniform(-limit_out,limit_out, (self.hidden_layer, n_output))
        self.BiasOutput = np.zeros((1, n_output))
     
    #Sigmoid Function
    def sigmoid(self, z):
        return 1.0 / (1.0 + np.exp(-z))
    
    #Sigmoid Derivative Function
    def sigmoid_derivative(self, z):
        return self.sigmoid(z) * (1 - self.sigmoid(z))
     
    #SoftMax Function (Output Layer)    
    def softmax(self, z):
        e_x = np.exp(z - np.max(z, axis=-1, keepdims=True))
        return e_x / np.sum(e_x, axis=-1, keepdims=True)
    
    #SoftMax Gradient Function
    def softmax_gradient(self, z):
        return self.softmax(z) * (1.0 - self.softmax(z))
    
    #Cross-Entropy Loss Function
    def loss(self, h, y):
        h = np.clip(h, 1e-15, 1 - 1e-15)
        return (-y * np.log(h) - (1 - y) * np.log(1 - h))
    
    #Cross-Entropy Loss Gradient Function
    def loss_gradient(self, h, y):
        h = np.clip(h, 1e-15, 1 - 1e-15)
        return -(h/y) + (1-h)/(1-y)
    
    #Accuracy Score Function
    def accuracy_score(self, y_true, y_pred):
        accuracy = np.sum(y_true == y_pred, axis=0) / len(y_true)
        return accuracy
    
    #Prediction Function
    def predict(self, X):
        hidden_input = np.array((X.dot(self.hiddenWeight) + self.BiasHidden),dtype=np.float32)
        hidden_output = self.sigmoid(hidden_input)
        output_layer_input = hidden_output.dot(self.outputWeight) + self.BiasOutput
        y_pred = self.softmax(output_layer_input)
        return y_pred
    
    #Fit Function
    def fit(self, X, y):
        self.initial_weights(X, y)
        n_epoch = 1
        
        while(n_epoch <= self.epoch):
            
            # Forward Propogation
            #hidden Layer
            hidden_input = np.array((X.dot(self.hiddenWeight) + self.BiasHidden),dtype=np.float32)
            hidden_output = self.sigmoid(hidden_input)
            #output layer
            output_layer_input = hidden_output.dot(self.outputWeight) + self.BiasOutput
            y_pred = self.softmax(output_layer_input)
            
            #Backward Propogation
            #Output Layer Gradient
            grad_out_input = self.loss_gradient(y, y_pred) * self.softmax_gradient(output_layer_input)
            grad_output = hidden_output.T.dot(grad_out_input)
            grad_biasoutput = np.sum(grad_out_input,axis=0,keepdims=True)
            #Hidden Layer Gradient
            grad_input_out = grad_out_input.dot(self.outputWeight.T) * self.sigmoid_derivative(hidden_input)
            grad_input = X.T.dot(grad_input_out)
            grad_biasinput = np.sum(grad_input_out, axis=0, keepdims=True)
            
            #Updating Weights
            self.outputWeight = self.outputWeight - (self.learning_rate * grad_output)
            self.BiasOutput = self.BiasOutput - (self.learning_rate *grad_biasoutput)
            self.hiddenWeight = self.hiddenWeight - (self.learning_rate * grad_input)
            self.BiasHidden = self.BiasHidden - (self.learning_rate * grad_biasinput)
                        
            
            n_epoch += 1
            
       
```

## Main Function

```python
def main():
    data = pd.read_csv("../input/heart-failure-prediction/heart.csv")
    df = data.sample(frac=1,random_state=32)
    X_label = df.drop(["HeartDisease"],axis=1)
    y_label = df["HeartDisease"]
    
    #Normalize X-Labels
    X1 = X_label.copy()
    X1 = (X1 - X1.mean())/X1.std() #Standardizing the X-Labels
    X1 = X1.to_numpy()
    
    #Categorize Y-Label One hot Encoding
    y1 = np.zeros((y_label.shape[0], (np.amax(y_label)+1)))
    y1[np.arange(y_label.shape[0]), y_label] = 1
    
    #Train-Test Split
    splitting_len = int(0.75 * len(df))
    X_train = X1[:splitting_len] 
    X_test  = X1[splitting_len:]
    y_train = y1[:splitting_len]
    y_test  = y1[splitting_len:]
    
    
    model = MultiLayerPerceptron(hidden_layer = 16, epoch=1000, learning_rate=0.01, verbose=True)
    model.fit(X_train, y_train)
    
    y_pred = np.argmax(model.predict(X_test), axis=1)
    y_test = np.argmax(y_test, axis=1)

    accuracy = model.accuracy_score(y_test, y_pred)
    print ("Accuracy:", accuracy)
    
    #with open('Bestfitmodel1.pickle', 'wb') as file:
    #pickle.dump(model, file)
```

```python
if __name__ == "__main__":
    main()
```
